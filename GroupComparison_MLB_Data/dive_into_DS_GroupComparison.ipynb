## Differences between Sample Data

```python
alldifferences=[]
for i in range(1000):
    newsample1=mlb.sample(n=30,random_state=i*2)
    newsample2=mlb.sample(n=30,random_state=i*2+1)
    alldifferences.append(newsample1['height'].mean()-newsample2['height'].mean())

print(alldifferences[0:10])
```

```python
import seaborn as sns
sns.set()
ax=sns.distplot(alldifferences).set_title("Differences Between Sample Means")
plt.xlabel('Difference Between Mean (Inches)')
plt.ylabel('Relative Frequency')
plt.show()
```

```python
largedifferences=[diff for diff in alldifferences if abs(diff) >= 1.6]
print(len(largedifferences))
```

```python
smalldifferences=[diff for diff in alldifferences if abs(diff) >= 0.6]
print(len(smalldifferences))
```

## The t-Test

### Hypothesis Testing
Hypothesis 0: sample1 and sample3 are randomly samples from the same population.
<br>
Hypothesis 1: sample1 and sample3 are not randomly samples from the same population.
<br>
1. Assume Hypothesis 0, the null hypothesis, is true.
<br>
2. Find how likely we are to observe sample means that differ by as much as our observed sample means, assuming that Null Hypothesis is true. The likelihood of this ocurring is called the p-value.
<br>
3. If the p-value is samll enough, we reject the Null Hypothesis, and we're therefore willing to accept Hypothesis 1. 

```python
import scipy.stats
scipy.stats.ttest_ind(sample1['height'],sample2['height'])
```

```python
scipy.stats.mannwhitneyu(sample1['height'],sample2['height'])
```

p-value on ttest is 0.283 and p-value on meanwhitneyu is 0.389 which is greater than 0.05. Since the p-value is high than these two samples come from the same population.

## Break down by Position

```python
Position_groupby = mlb.groupby('position')
```

```python
#Heights grouped by position
Starting_Pitcher_H = Position_groupby.get_group('Starting_Pitcher')['height']
Relief_Pitcher_H = Position_groupby.get_group('Relief_Pitcher')['height']
Catcher_H = Position_groupby.get_group('Catcher')['height']
First_Baseman_H = Position_groupby.get_group('First_Baseman')['height']
Second_Baseman_H = Position_groupby.get_group('Second_Baseman')['height']
Third_Baseman_H = Position_groupby.get_group('Third_Baseman')['height']
Shortstop_H = Position_groupby.get_group('Shortstop')['height']
Outfielder_H = Position_groupby.get_group('Outfielder')['height']
DH_H = Position_groupby.get_group('Designated_Hitter')['height']

#Data Visualization for height based on position
plt.boxplot([Starting_Pitcher_H,Relief_Pitcher_H,Catcher_H,First_Baseman_H,Second_Baseman_H,Third_Baseman_H,Shortstop_H,Outfielder_H,DH_H])
plt.xlabel('Position Number')
plt.ylabel('Height (Inches)')
plt.title('MLB Heights based on Position')
plt.xticks([1,2,3,4,5,6,7,8,9],['S 1','R 1','2','3','4','5','6','7-9','DH'])
plt.show()
```
